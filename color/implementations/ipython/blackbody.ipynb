# Blackbody

A [blackbody](http://en.wikipedia.org/wiki/Black_body) is an hypothetical perfect absorber and radiator of energy.

A *blackbody* in [thermal equilibrium](http://en.wikipedia.org/wiki/Thermal_equilibrium) emits electromagnetic radiation called [blackbody radiation](http://en.wikipedia.org/wiki/Black-body_radiation).

The spectral radiance of a blackbody at thermodynamic temperature $T [K]$ in a medium having index of refraction $n$ is given by the [Planck's law](http://en.wikipedia.org/wiki/Planck%27s_law) equation:

$$
L_{e\lambda}(\lambda,T)=\cfrac{C_1n^{-2}\lambda^{-5}}{\pi}{\Biggl[\exp\biggl(\cfrac{C_2}{n\lambda T}\biggr)-1\Biggr]^{-1}}
$$
where
$$
C_1 = 2\pi hc^2\\
C_2=\cfrac{hc}{k}
$$

$h$ is Planck's constant, $c$ is the speed of light in vacuum, $k$ is the Boltzmann constant and $\lambda$ is the wavelength.

As per *CIE publication CIE 15:2004 "Colorimetry"* recommendation $C_2$ value when used in colorimetry should be $C_2= 1,4388 x 10^{-2} m K$ as defined by the International Temperature Scale (ITS-90).

$C_1$ value is given by the Committee on Data for Science and Technology (CODATA) and should be $C_1 = 3,741771 x 10^{16} W m^2$.

In the current *CIE publication CIE 15:2004 "Colorimetry"* recommendation, color temperature and correlated color temperature are calculated with $n = 1$.

> Reference: https://law.resource.org/pub/us/cfr/ibr/003/cie.15.2004.pdf

[Color](https://github.com/KelSolaar/Color/) implements some *blackbody* related objects available in *color.spectral.blackbody* sub-package:

```python
import color.spectral.blackbody
```

> Note: *color.spectral.blackbody* sub-package content is also available in *color* namespace.

The *Planck's law* is called using either the *color.planck_law* or  *color.blackbody_spectral_radiance* definitions, they are expecting the wavelength $\lambda$ to be in nanometers and the temperature $T$ to be in degree kelvin:

```python
import color

color.planck_law(500 * 1e-9, 5500)
```

Generating the spectral power distribution of a *blackbody* is done using the *color.blackbody_spectral_power_distribution* definition:

```python
color.blackbody_spectral_power_distribution(6500, 0, 10000, 10)
```

With its temperature lowering, the blackbody peak shifts to longer wavelengths while its intensity also decreases:

```python
from color.implementations.matplotlib.plots import *

# Plotting various *blackbodies* spectral power distributions.
blackbodies_spds = [color.blackbody_spectral_power_distribution(i, 0, 10000, 10) for i in range(1000, 15000, 1000)]
multi_spectral_power_distribution_plot(blackbodies_spds,
                                       y_label=u"W / (sr m²) / m",
                                       use_spds_colors=True,
                                       normalize_spds_colors=True,
                                       legend_location="upper right",
                                       bounding_box=[0, 1000, 0, 2.25e15])
```
